```python
import wandb

from zoology.analysis.utils import fetch_wandb_runs

# Replace with your wandb username and project name
project_path = "gpt6/zoology"

# Initialize the wandb API
api = wandb.Api()

df = fetch_wandb_runs(
    launch_id=[
        "default-2024-01-12-00-51-08",
    ], 
    project_name="zoology"
)

# df.head()

# with open('columns.txt', 'w') as file:
#     file.write(','.join(df.columns))
```

```python
import datetime
import pandas as pd

# Assuming 'df' is a DataFrame with your data

# Drop rows with NaN values in the '_timestamp' column
df = df.dropna(subset=['_timestamp'])

# Convert timestamps to datetime objects
df['_timestamp'] = df['_timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x))

# Calculate the earliest and latest timestamps
earliest_timestamp = df['_timestamp'].min()
latest_timestamp = df['_timestamp'].max()

# Calculate average runtime
average_runtime = df['_runtime'].mean()

# Calculate shortest and longest runtimes
shortest_runtime = df['_runtime'].min()
longest_runtime = df['_runtime'].max()

# Print results
print(f"Earliest Timestamp: {earliest_timestamp}")
print(f"Latest Timestamp: {latest_timestamp}")
print(f"Average Runtime: {average_runtime} (units depend on the dataset)")
print(f"Shortest Runtime: {shortest_runtime} (units depend on the dataset)")
print(f"Longest Runtime: {longest_runtime} (units depend on the dataset)")
```
